```python
from sklearn.datasets import load_digits
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
```

```python
#Load the hand written dataset
digits = load_digits()
data = digits.data
labels = digits.target
```

## LDA

```python
#Performing LDA
lda = LinearDiscriminantAnalysis(n_components=2)
lda_results = lda.fit_transform(data, labels)
```

```python
#Plot LDA
plt.figure(figsize=(10,8))

#Loop through each class label
for label in set(labels):
    indices = labels == label
    plt.scatter(lda_results[indices, 0], lda_results[indices, 1], label=label, cmap='viridis')
    
plt.title('LDA Visualization of Handwritten Digits')
plt.legend()
plt.colorbar()

#Calculate evaluation metrics
sil_score = silhouette_score(data, labels)
daview_score = davies_bouldin_score(data, labels)


plt.text(0.05, -0.3, f"Original Dimension: {data.shape[1]} \n Reduced Dimension: {lda_results.shape[1]} \nSilhoutte Score: {sil_score:.4f} \nDaview Score: {daview_score:.4f} \nReduction in %: {((1-(lda_results.shape[1])/(data.shape[1]))*100):.2f}%", transform=plt.gca().transAxes)

plt.show()
```

## SVD

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
```

```python
#Load the hand written dataset
digits = load_digits()
data = digits.data
labels = digits.target
```

```python
#Perform SVD
svd = TruncatedSVD(n_components=2, random_state=43)
svd_results = svd.fit_transform(data)
```

```python
#Plot SVD
plt.figure(figsize=(10,8))

#Loop through each class label
for label in set(labels):
    indices = labels == label
    plt.scatter(svd_results[indices, 0], svd_results[indices, 1], label=label, cmap='viridis')
    
plt.title('SVD Visualization of Handwritten Digits')
plt.legend()
plt.colorbar()

#Calculate evaluation metrics
sil_score = silhouette_score(data, labels)
daview_score = davies_bouldin_score(data, labels)


plt.text(0.05, -0.3, f"Original Dimension: {data.shape[1]} \n Reduced Dimension: {svd_results.shape[1]} \nSilhoutte Score: {sil_score:.4f} \nDaview Score: {daview_score:.4f} \nReduction in %: {((1-(svd_results.shape[1])/(data.shape[1]))*100):.2f}%", transform=plt.gca().transAxes)

plt.show()
```
