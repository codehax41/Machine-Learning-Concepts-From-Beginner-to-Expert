```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def distancess(p1, p2):
    return np.sqrt(np.sum(p1 - p2)**2)
```

```python
def generate_points(num_points, num_dimensions):
    points = np.random.rand(num_points, num_dimensions)
    return points
```

```python
def plot_distance(num_points, num_dimensions):
    average_dist = []
    
    for num_dimensions in range(1, 100):
        points = generate_points(num_points, num_dimensions)
        
        distance = []
        for i in range(num_points):
            for j in range(num_points):
                if i!=j:
                    distance.append(distancess(points[i], points[j]))
                    
        average_dist.append(distance)
        
    plt.plot(range(1,100), average_dist)
    plt.xlabel('Number of Dimensions')
    plt.ylabel("Average Distance")
    plt.title("Curse of Dimentionality")
    plt.show()
```

```python
if __name__ == '__main__':
    plot_distance(100, 9)
```
