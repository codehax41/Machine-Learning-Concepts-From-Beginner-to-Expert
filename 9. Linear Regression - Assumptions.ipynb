## Assumption 1: Linearity

```python
import scipy.stats as stats
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 10, 100)
y = 2 * x + 3+ np.random.normal(0,1,100)
```

```python
plt.scatter(x, y)
plt.title("Assumption 1: Linearity")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Assumption 2: Independence of Errors

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
x = np.linspace(1, 10, 100)
y = 2 * x + 3+ np.random.normal(0,1,100)
y[50:] += np.arange(1, 51) #Autocorrelation
```

```python
y
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
```

```python
plt.scatter(model.predict(), model.resid)
plt.title("Assumption 2: Independence of Errors")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

```python
#without autocorrelation
```

```python
x = np.linspace(1, 10, 100)
y = 2 * x + 3+ np.random.normal(0,1,100)


x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

plt.scatter(model.predict(), model.resid)
plt.title("Assumption 2: Independence of Errors")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

## Assumption 3: Homoscedasticity

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
x = np.linspace(1, 10, 100)
y = 2 * x + 3 + x * np.random.normal(0,1,100) #Introduce Hetroscedasity
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
```

```python
plt.scatter(model.predict(), model.resid)
plt.title("Assumption 3: Homoscedasticity")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

## Assumption 4: Normality of Error

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
x = np.linspace(1, 10, 100)
y = 2 * x + 3 + np.random.uniform(0,1,100) #Introducing non-normality
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
```

```python
residuals = model.resid
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Assumption 4: Normality of Error")
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = 2 * x + 3 + np.random.normal(0,1,100)

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

residuals = model.resid
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Assumption 4: Normality of Error")
plt.show()
```
