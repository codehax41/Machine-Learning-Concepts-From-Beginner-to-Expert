```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
market_df = pd.read_csv('market_fact.csv')
order_df = pd.read_csv('orders_dimen.csv')
cust_df = pd.read_csv('cust_dimen.csv')
prod_df = pd.read_csv('prod_dimen.csv')
```

```python
sns.boxplot(market_df['Sales'])
```

```python
sns.boxplot(market_df['Sales'])
plt.yscale('log')
plt.show()
```

```python
market_df['Sales'].describe()
```

```python
df1 = pd.merge(market_df, prod_df, on = "Prod_id", how='inner')
```

```python
df1.head()
```

```python
sns.boxplot(x='Product_Category', y='Sales', data=df1)
plt.show()
```

```python
sns.boxplot(x='Product_Category', y='Sales', data=df1)
plt.yscale('log')
plt.show()
```

```python
sns.boxplot(x='Product_Category', y='Profit', data=df1)
plt.show()
```

```python
df2 = df1[(df1.Profit<1000) & (df1.Profit>-1000)]
```

```python
sns.boxplot(x='Product_Category', y='Profit', data=df2)
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.subplot(1, 2, 1)
sns.boxplot(x='Product_Category', y='Sales', data=df2)
plt.title('Sales')
plt.yscale('log')


plt.subplot(1, 2, 2)
sns.boxplot(x='Product_Category', y='Profit', data=df2)
plt.title('Profit')

plt.show()
```

```python
df3 = pd.merge(df2, cust_df, how='inner', on='Cust_id')
```

```python
sns.boxplot(x="Customer_Segment", y = 'Profit', data= df3)
plt.show()
```

```python
sns.boxplot(x="Customer_Segment", y = 'Profit', hue="Product_Category",  data= df3)
plt.show()
```

```python
sns.boxplot(x=df3['Product_Category'], y = 100*(df3['Shipping_Cost']/df3['Sales']))
plt.show()
```

```python
sns.barplot(x='Product_Category', y='Sales', data=df3)
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.subplot(1, 2, 1)
sns.barplot(x='Product_Category', y='Sales', data=df2)
plt.title('Average Sale')

plt.subplot(1, 2, 2)
sns.barplot(x='Product_Category', y='Sales', data=df2, estimator=np.median)
plt.title('Median Sale')

plt.show()
```

```python
sns.barplot(x='Product_Category', y='Sales', hue="Customer_Segment", data=df3, estimator=np.median)
```

```python
sns.barplot(x='Product_Category', y='Profit', hue="Customer_Segment", data=df3, estimator=np.median)
```

```python
sns.barplot(x='Customer_Segment', y='Profit', hue="Product_Category", data=df3, estimator=np.median)
```

```python
sns.barplot(x='Profit', y="Product_Sub_Category", data=df3, estimator=np.median)
```

```python
sns.countplot(y='Product_Sub_Category', data=df3)
```
