## Cost Function

```python
import numpy as np
```

```python
actual_values = np.array([2.5, 3.0, 1.8, 4.0, 5.2])
predicted_values = np.array([2.7, 2.9, 1.5, 3.9, 5.0])
```

```python
def mean_squared_error(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()
```

```python
mse = mean_squared_error(actual_values, predicted_values)
```

```python
print("Mean Squared Error", mse)
```

```python
actual_values = np.array([2.5, 3.0, 1.8, 4.0, 5.2])
predicted_values = np.array([2.7, 2.9, 10, 40, 50])
```

```python
mse = mean_squared_error(actual_values, predicted_values)
```

```python
print("Mean Squared Error", mse)
```

## Gradient Decent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(theta):
    return theta**2 + 2*theta + 1
```

```python
def gradient(theta):
    return 2*theta + 2
```

```python
theta = 3.0
```

```python
lr = 0.1
num_iteraion = 100
```

```python
theta_history = []
cost_history = []
```

```python
for i in range(num_iteraion):
    gradient_val = gradient(i)
    theta -= lr * gradient_val
    theta_history.append(theta)
    cost_history.append(cost_function(theta))
    
print("Minimum at thete =", theta)
print("Cost at minimum:", cost_function(theta))

plt.figure(figsize = (10, 6))
plt.plot(theta_history, cost_history, marker='o', linestyle='-', color='b')
plt.title("Gradient Decent")
plt.xlabel("Theta")
plt.ylabel("Cost")
plt.grid()
plt.show()
```
