```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
```

```python
#Function to create synthetic data with multiple clusters
def generate_synthetic_data():
    data, _ = make_blobs(n_samples=300, centers=4, random_state=42)
    return data
```

```python
data = generate_synthetic_data()
```

```python
def elbow_method(data, max_k):
    distortions = []
    for k in range(1, max_k+1):
        kmeans = KMeans(n_clusters=k, random_state=0)
        kmeans.fit(data)
        distortions.append(kmeans.inertia_)
    return distortions
```

```python
def silhouette_method(data, max_k):
    silhouette_ = []
    for k in range(2, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=0)
        labels = kmeans.fit_predict(data)
        silhouette_.append(silhouette_score(data, labels))
    return silhouette_
```

```python
def plot_elbow_method(data, max_k):
    distortions = elbow_method(data, max_k)
    plt.plot(range(1, max_k+1), distortions, marker='o')
    plt.title('Elbow Method for Optimal k')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Distortion (Intertia)')
    plt.show()
```

```python
def plot_silhouette_method(data, max_k):
    silhouette_sc = silhouette_method(data, max_k)
    plt.plot(range(2, max_k+1), silhouette_sc, marker='o')
    plt.title('Silhouette Method for Optimal k')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Silhouette Score')
    plt.show()
```

```python
#Choose a max value for k
max_k = 10
```

```python
plot_elbow_method(data, max_k)
plot_silhouette_method(data, max_k)
```
