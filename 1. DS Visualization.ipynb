```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("market_fact.csv")
```

```python
df.head()
```

```python
sns.distplot(df['Shipping_Cost'])
```

```python
sns.distplot(df['Shipping_Cost'], hist=False)
```

```python
sns.distplot(df['Shipping_Cost'], rug=True)
```

```python
plt.subplot(2,2,1)
plt.title("Y is log of X")
sns.distplot(df['Shipping_Cost'])

plt.subplot(2,2,2)
plt.title("log of X**2")
sns.distplot(df['Shipping_Cost'])

plt.subplot(2,2,3)
plt.title("Y is log of X")
sns.distplot(df['Shipping_Cost'])

plt.subplot(2,2,4)
plt.title("log of X**2")
sns.distplot(df['Shipping_Cost'])
plt.show()
```

```python
#Box plot
sns.boxplot(df['Order_Quantity'])
```

```python
#Multivariate Dist Plot
sns.jointplot(x='Sales',y='Profit', data=df)
plt.show()
```

```python
df1 = df[(df['Profit']<10000) & (df['Sales']<20000)]
```

```python
#Multivariate Dist Plot
sns.jointplot(x='Sales',y='Profit', data=df1)
plt.show()
```

```python
df2 = df[(df['Profit']<100) & (df['Profit']>-100) & (df['Sales']<200)]
sns.jointplot(x="Sales", y='Profit', data = df2, color="g")
```

```python
sns.jointplot(x="Sales", y='Profit', data = df2, color="g", kind='hex')
```

```python
sns.jointplot(x="Sales", y='Profit', data = df2, color="g", kind='kde')
```

```python
sns.pairplot(df)
```

```python
#Corr
corr = df.corr()
corr1 = round(corr, 3)
```
